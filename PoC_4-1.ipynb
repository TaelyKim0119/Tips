```python

import re
def clean_text(text):
    text = re.sub(r"http\S+", "", str(text))  # URL 제거
    text = re.sub(r"[^\uAC00-\uD7A3a-zA-Z0-9\s]", "", text)  # 특수문자 제거
    return text.strip().lower()

def extract_keywords(text):
    tokens = okt.nouns(text)
    tokens = [t for t in tokens if len(t) > 1]  # 길이 필터
    return tokens

# 적용
for col in ['title', 'description']:
    df[col] = df[col].astype(str).map(clean_text)
    df[col + '_tokens'] = df[col].map(extract_keywords)

# 토큰 합치기
df['merged_tokens'] = df['title_tokens'] + df['description_tokens']
```

```python
from collections import Counter

# merged_tokens에서 모든 키워드를 평탄화(flatten)해서 하나의 리스트로 만들기
all_tokens = [token for tokens in df['merged_tokens'] for token in tokens]

# 키워드 빈도 수 계산
keyword_counts = Counter(all_tokens)

# 상위 300개 키워드 추출
top_300_keywords = [kw for kw, count in keyword_counts.most_common(300)]
```

```python
# 불용어 리스트 정의 (앞서 수집한 의미 약한 단어들)
stopwords = [
    '이유','문의','방법','지금','오늘','이번','모든','가장','이후','최근','가운데',
    '지난','시작','정리','생각','때문','전체','가능','반드시','혹시','반응',
    '생중계','시청','구독','면제','일부','현재','머그','젤리','초코','테디',
    '프롬','프트','오거','이돈','유영상','뉴스','내용','영상','자막'
]

# 렉시콘 후보에서 불용어 제거
filtered_keywords = [kw for kw in top_300_keywords if kw not in stopwords]
```

```python
from collections import Counter, defaultdict
import pandas as pd
import math

# 1. merged_tokens 가져오기
docs = df["merged_tokens"]

# 2. 전체 문서 수
N = len(docs)

# 3. 단어 등장 문서 수
word_doc_freq = Counter()
co_occurrence = Counter()

for tokens in docs:
    tokens_set = set(tokens)
    if "skt" not in tokens_set:
        continue
    for word in tokens_set:
        word_doc_freq[word] += 1
    for neg_kw in negative_lexicon:
        if neg_kw in tokens_set:
            co_occurrence[neg_kw] += 1

# SKT 등장 문서 수
skt_doc_count = word_doc_freq["skt"]
```

```python
skt_doc_count
```

```python
def compute_pmi(skt_count, word_count, co_count, total_docs):
    p_xy = co_count / total_docs
    p_x = skt_count / total_docs
    p_y = word_count / total_docs
    if p_xy == 0:
        return 0
    return round(math.log2(p_xy / (p_x * p_y)), 3)

# PMI 계산 결과 저장
pmi_results = []

for neg_kw in negative_lexicon:
    co_count = co_occurrence[neg_kw]
    word_count = word_doc_freq[neg_kw]

    if co_count == 0:
        continue

    pmi_score = compute_pmi(skt_doc_count, word_count, co_count, N)

    pmi_results.append({
        "부정 키워드": neg_kw,
        "skt+부정 키워드 동시 등장 문서 수": co_count,
        "skt 등장 문서 수": skt_doc_count,
        "부정 키워드 등장 문서 수": word_count,
        "PMI": pmi_score
    })

# 정렬
pmi_df = pd.DataFrame(pmi_results).sort_values(by="PMI", ascending=False).reset_index(drop=True)
```
