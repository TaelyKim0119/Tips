```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime



# 2. 날짜 처리
df["published"] = pd.to_datetime(df["published"]).dt.tz_localize(None)
df["date"] = df["published"].dt.date

# 3. 기준일 설정 (이슈 발생일)
event_date = datetime(2024, 4, 22)
df["days_after_event"] = (df["published"] - event_date).dt.days.clip(lower=0)

# 4. W 수식 정의
def compute_W(row, lambda_=0.7, beta=0.05):
    M = np.log1p(row["commentCount"] + row["likeCount"] + row["viewCount"])
    S = row["final_negative_emotion_score"]
    decay = np.exp(-beta * row["days_after_event"])
    return np.log1p(M * (1 + lambda_ * S)) * decay

df["W_score"] = df.apply(compute_W, axis=1)

# 5. 로그 스케일 열 추가 (EDA용)
df["log_view"] = np.log1p(df["viewCount"])
df["log_like"] = np.log1p(df["likeCount"])
df["log_comment"] = np.log1p(df["commentCount"])

# 6. 날짜별 평균 계산
trend_log = df.groupby("date")[["log_view", "log_like", "log_comment"]].mean().reset_index()
w_by_date = df.groupby("date")["W_score"].mean().reset_index()

# 7. 병합
plot_df = pd.merge(trend_log, w_by_date, on="date", how="left")

# 8. 시각화
plt.figure(figsize=(12, 6))

plt.plot(plot_df["date"], plot_df["W_score"], label="W 점수", marker="o", linestyle="--", color="black")

plt.title("📈 통합 로그 변화 + W 점수 추이")
plt.xlabel("게시 날짜")
plt.ylabel("로그/영향력 점수")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 2. 날짜 처리 및 필터링
df["published"] = pd.to_datetime(df["published"]).dt.tz_localize(None)
df["date"] = df["published"].dt.date
mask = (df["published"] >= "2025-04-01") & (df["published"] <= "2025-05-31")
df = df[mask].copy()

# 3. 로그 스케일 언급량 계산
df["log_view"] = np.log1p(df["viewCount"])
df["log_like"] = np.log1p(df["likeCount"])
df["log_comment"] = np.log1p(df["commentCount"])
df["log_M"] = df[["log_view", "log_like", "log_comment"]].sum(axis=1)

# 4. 날짜별 집계 (log_M 총합, 부정 감정 평균)
daily_agg = df.groupby("date").agg({
    "log_M": "sum",
    "final_negative_emotion_score": "mean"
}).sort_index().reset_index()

# 5. Δlog(Mt) 계산 및 W 점수 수식 적용
lambda_ = 1
daily_agg["delta_log_M"] = daily_agg["log_M"].diff()
daily_agg["W_score"] = daily_agg["delta_log_M"] * (1 + lambda_ * daily_agg["final_negative_emotion_score"])

# 6. 시각화
fig, ax1 = plt.subplots(figsize=(12, 5))

# W 점수 그래프 (왼쪽 y축)
ax1.plot(daily_agg["date"], daily_agg["W_score"], marker="o", linestyle="--", color="black", label="W 점수")
ax1.set_xlabel("날짜")
ax1.set_ylabel("W 점수", color="black")
ax1.tick_params(axis="y", labelcolor="black")

# log_M 그래프 (오른쪽 y축)
ax2 = ax1.twinx()
ax2.plot(daily_agg["date"], daily_agg["log_M"], marker="s", linestyle="-", color="red", label="log_M")
ax2.set_ylabel("log_M", color="red")
ax2.tick_params(axis="y", labelcolor="red")

plt.title("날짜별 W 점수와 log_M 추이")
plt.xticks(rotation=45)

# 범례 통합
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

plt.grid(True)
plt.tight_layout()
plt.show()

# 7. 최종 데이터 확인
pd.set_option('display.unicode.east_asian_width', True)
print(daily_agg.tail())
```
